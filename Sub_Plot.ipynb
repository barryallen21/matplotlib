```python
from matplotlib import pyplot as plt
import numpy as np
import random
from matplotlib import style as s
import pandas as pd
```

```python
plt.figure(figsize=(16,9))
Days = [1,2,3,4,5]
Delhi_Temp = [78,12,45,32,67]
Bombay_Temp = [23,12,67,90,65]

plt.subplot(3,2,1)
plt.plot(Days,Delhi_Temp , marker = 'o', linestyle = ':' , color = 'm' , linewidth = 2 , markersize = 7 ,
         label = "Delhi Temp")
plt.plot(Days,Bombay_Temp , marker = 'o', linestyle = ':' , color = 'k' , linewidth = 2 , markersize = 7 ,
         label = "Bombay Temp")
plt.axis([0,6,0,120])
plt.title("Temp of first five Days",fontsize = 20 , color = "y" , loc = 'left')
plt.xlabel("Days->",fontsize = 20 , loc = 'right', color = "k")
plt.ylabel("Temp->",fontsize = 20, loc = 'top', color = "k")
plt.legend(loc = 2 )
plt.grid(which = 'major')

plt.subplot(3,2,2)
java_users = np.random.randint(25,50 , (100))
python_users = np.random.randint(25,50 , (100))
bins = [25,30,35,40,45,50]
plt.hist([java_users,python_users] , bins ,color = ['m','c'], rwidth = .9 , histtype='bar', orientation='vertical' 
, label = ['java users','python users'])
#plt.hist(python_users , bins ,color = 'c', rwidth = .9 , histtype='bar', orientation='vertical' 
#, label = 'python users')
plt.title("JAVA USERS & python users" , fontsize = 20 )
plt.xlabel("age of user") 
plt.ylabel("no of users")
plt.legend(loc=2)
plt.grid(color ='k')

plt.subplot(3,2,3)
Classes = ['python','c#','java','c++','js']
class1_stds = np.random.randint(10,50 ,(5))
class2_stds = np.random.randint(10,50 ,(5))
class3_stds = np.random.randint(10,50 ,(5))
class_index = np.arange(len(Classes))
width = .25
#plt.barh() used to drew the bar chaet horizontally
plt.bar(class_index,class1_stds , width,align = 'center' , color = 'm' , edgecolor = 'k' , linewidth =3,
       alpha = .9 , linestyle = '-' , label = 'class1 studs')
# 'visible =False'used to hide chart
plt.bar(class_index + width,class2_stds ,width, align = 'center' , color = 'c' , edgecolor = 'k' , linewidth =3,
       alpha = .9 , linestyle = '-' , label = 'class2 studs')
plt.bar(class_index + 2*width,class2_stds ,width, align = 'center' , color = 'r' , edgecolor = 'k' , linewidth =3,
       alpha = .9 , linestyle = '-' , label = 'class3 studs')
plt.title("Bar Chart of dummy class" ,fontsize = 20 , color='b')
plt.xlabel('class ->' , color = 'k',fontsize = 15 )
plt.ylabel("no of std ->" , color ='k',fontsize = 15)
plt.xticks(class_index,Classes)
plt.legend()
plt.grid()

plt.subplot(3,2,4)
df_1 = pd.read_csv("C:\\dumyexcel\\ExportCSV (2).csv")
df_1
emp_ID = df_1['ID']
emp_age = df_1['Scrap']
emp_title = df_1['Job Title']
plt.scatter(emp_ID,emp_age, color='k',alpha=.9 ,edgecolor ='r',label='emp age',marker='o',s=50 ,linewidth = .8)
plt.scatter(emp_title,emp_age, color='k',alpha=.9 ,edgecolor ='m',label='emp title',marker='o',s=50 ,linewidth = .8)
plt.title("Employee age scatter plot",fontsize=20 ,color = 'y')
plt.xlabel('ID->',fontsize=20)
plt.ylabel('AGE->',fontsize=20)
plt.grid()
plt.legend()

plt.subplot(3,2,5)
languages = ['python','c#','c++','java','js']
class_std = [12,45,50,34,23]
explode =[0,0,0,.2,0]
color =['c','g','b','m','r']
textprops = {'fontsize':10}
wedgeprops = {'linewidth':3,'edgecolor':'k'}
plt.pie(class_std , labels =languages ,explode=explode ,colors=color,autopct = '%.2f%%',shadow = True,
        radius=1.5 ,startangle = 270 ,textprops=textprops,pctdistance=.7 ,labeldistance =1.09,
       counterclock =False ,wedgeprops=wedgeprops)
plt.legend(loc=10)
plt.subplot(3,2,6,projection = 'polar' ,facecolor = 'k')#frameon used is to hide a plot

plt.show()
```
